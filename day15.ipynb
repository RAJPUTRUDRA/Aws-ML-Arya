```python
# handle missing values
import numpy as np
import pandas as pd
data = {
    "experience":[1,np.nan,3,4,5],
    "salary":[100,200,300,400,np.nan],
    "name":['hello','dheeraj','Nikhil','Raj','vinay']
}
df = pd.DataFrame(data)
print(df)
# null value filter | find out
print(df.isnull())

# sum of null
print(df.isnull().sum())

# dropna | drop complete row
df.dropna()
print(df)

# fillna | null -> fill
df.fillna(5)
```

```python
import matplotlib.pyplot as plt
#pie chart
x = ["apple","banana","carrot","mango"]
y = [30,50,70,5]
fig, aux = plt.subplots(2,2)
aux[0,0].pie(y,labels=x)

# line chart
data = {
    "experience":[1,2,3,4,5],
    "salary":[100,200,300,400,500]
}
x = data['experience']
y = data['salary']
aux[0,1].plot(x,y)


#bar chart
x = ["apple","banana","carrot","mango"]
y = [30,50,70,5]

aux[1,0].bar(x,y)

#multiline chart
x = ["apple","banana","carrot","mango"]
y = [30,50,7,1]
aux[1,1].plot(x,y)
plt.tight_layout()
plt.show()
```
