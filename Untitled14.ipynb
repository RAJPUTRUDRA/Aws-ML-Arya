```python
# operations on rows
import pandas as pd
data = {
    "Emp_ID": [101, 102, 103, 104, 105, 106],
    "Name": ["Amit", "Riya", "Raj", "Sara", "John", "Neha"],
    "Department": ["IT", "HR", "Finance", "IT", "Sales", "HR"],
    "Salary": [50000, 45000, 60000, 55000, 48000, 52000],
    "Experience": [2, 3, 5, 4, 1, 3]
}
df = pd.DataFrame(data)
print(df)
#filter one generator
df_filtered = df[df["Salary"] > 50000]
print(df_filtered)

#filter one greater
print(df[df['Emp_ID'] > 104])
#106
print(df[df['Emp_ID'] == 106])
```

```python
# operations on rows
import pandas as pd
data = {
    "Emp_ID": [101, 102, 103, 104, 105, 106],
    "Name": ["Amit", "Riya", "Raj", "Sara", "John", "Neha"],
    "Department": ["IT", "HR", "Finance", "IT", "Sales", "HR"],
    "Salary": [50000, 45000, 60000, 55000, 48000, 52000],
    "Experience": [2, 3, 5, 4, 1, 3]
}
df = pd.DataFrame(data)
print(df)

#soring ascending
print(df.sort_values("Salary",ascending=True))

#soring descending
print(df.sort_values("Salary",ascending=False))

#two cols
print(df.sort_values(["Salary","Experience"],ascending=False))
```

```python
import pandas as pd
url = "https://raw.githubusercontent.com/rajendra0968jangid/Ds-Arya/main/student-data.json"
df = pd.read_json(url)
print(df)

print(df['gender'])
print(df[["name","physics"]])
df['physics']= df['physics'] + 10
print(df['physics'])

df.drop("gender",axis=1,inplace=True)
print(df)

print(df.loc[1,"name"])
print(df.iloc[1,1])

print(df.loc[[0,1,2,3]])
print(df.iloc[[1,2,3]])
print(df.iloc[[0,1],[0,1]])

print(df[df['physics'] > 90])
print(df[df['name'] == 'raj'])
print(df.sort_values("physics",ascending=False))

```
